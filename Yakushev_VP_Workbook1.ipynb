4.3.2 –ó–∞–¥–∞–Ω–∏–µ
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è. –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ (ùë•,–Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì –ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
import numpy as np
from numpy import sqrt, exp, divide, cos, log, abs, sin
from matplotlib import pyplot as plt


def f(x):
    up = sqrt(1 + exp(sqrt(x)) + cos(x ** 2))
    low = abs(1 - sin(x) ** 3)
    return divide(up, low) + log(abs(2 * x))


x = np.arange(1, 10, 0.5)
y = f(x)
plt.grid()
plt.plot(x, y)
plt.show()

x = x[0:int(len(x) / 2)]
y = f(x)
plt.grid()
plt.scatter(x, y)
plt.show()
```

4.3.3 –ó–∞–¥–∞–Ω–∏–µ
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(), –≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.

```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import exp, cos, log, abs, trapz


def f(x):
    return abs(cos(x * exp(cos(x) + log(x + 1))))


x = np.arange(0, 10, 1)
y = f(x)

plt.grid()
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()

area = trapz(y)
print('area:', area)
```

4.3.5 –ó–∞–¥–∞–Ω–∏–µ
–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è –¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: e^(x+y), sin(x + y) , cos(x + y) , x^y.

```python
from math import exp, sin, cos


def add(a, b):
    return a + b


def subtract(a, b):
    return a - b


def divide(a, b):
    if b == 0:
        raise ZeroDivisionError("–ù–∞ –Ω–æ–ª—å –¥–µ–ª–∏—Ç—å –Ω–µ–ª—å–∑—è")
    return a / b


def multiply(a, b):
    return a * b


def expf(a, b):
    return exp(a + b)


def sinf(a, b):
    return sin(a + b)


def cosf(a, b):
    return cos(a + b)


def powf(a, b):
    return a ** b


def helpf():
    return '–î–æ—Å—Ç—É–ø–Ω—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:\n' \
           '–ü–æ–º–æ—â—å - help\n' \
           '–ë–∞–∑–æ–≤—ã–µ: +, -, /, *\n' \
           '–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ: sin - sin(a+b), cos - cos(a+b), exp - e^(a+b)\n'


def calc(a, b, op):
    pattern = "{} {} {} = {}"
    if op == '+':
        return pattern.format(a, op, b, add(a, b))
    elif op == '-':
        return pattern.format(a, op, b, subtract(a, b))
    elif op == '*':
        return pattern.format(a, op, b, multiply(a, b))
    elif op == '/':
        return pattern.format(a, op, b, divide(a, b))
    elif op == 'exp':
        return "e^({} + {}) = {}".format(a, b, expf(a, b))
    elif op == 'sin':
        return "sin({} + {}) = {}".format(a, b, sinf(a, b))
    elif op == 'cos':
        return 'cos({} + {}) = {}'.format(a, b, cosf(a, b))
    elif op == 'help':
        return helpf()
    else:
        return None


def main():
    print(helpf())
    while True:
        operation = str(input("–í–≤–µ–¥–∏—Ç–µ –æ–ø–µ—Ä–∞—Ü–∏—é:"))
        if operation == 'exit':
            break
        a = float(input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ:"))
        b = float(input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ:"))
        print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", calc(a, b, operation))


if __name__ == "__main__":
    main()
```
